```python
import numpy as np

# Create a 2x2 ndarray representing the chessboard
chessboard = np.zeros((2, 2), dtype=int)

# Fill the chessboard with wheat, doubling the number of grains for each square
grains = 1  # Start with 1 grain on the first square
for i in range(chessboard.shape[0]):
    for j in range(chessboard.shape[1]):
        chessboard[i, j] = grains
        grains *= 2  # Double the number of grains for the next square

print("Original 2x2 Chessboard:")
print(chessboard)

# Transform the shape of the ndarray to a 1D array
flattened_chessboard = chessboard.flatten()
print("\nFlattened Chessboard:")
print(flattened_chessboard)

# Transform the shape back to 2x2
reshaped_chessboard = flattened_chessboard.reshape((2, 2))
print("\nReshaped Chessboard:")
print(reshaped_chessboard)
```

```python
import numpy as np

def chessboard_grains(n, m):
    """
    Create an ndarray containing the number of grains on an nxm chessboard.

    Args:
    - n (int): Number of rows on the chessboard.
    - m (int): Number of columns on the chessboard.

    Returns:
    - ndarray: An nxm array with the number of grains on each square.
    """
    # Create a flat array with the doubling pattern
    total_squares = n * m
    grains = np.array([2**i for i in range(total_squares)])
    
    # Reshape to an nxm array
    chessboard = grains.reshape((n, m))
    return chessboard

# Use the function to find the number of grains on an 8x8 chessboard
chessboard_8x8 = chessboard_grains(8, 8)
print("8x8 Chessboard with Grains:")
print(chessboard_8x8)

# Calculate the total number of grains on the chessboard
total_grains = chessboard_8x8.sum()
print(f"\nTotal number of grains on an 8x8 chessboard: {total_grains}")
```

```python
import numpy as np

def chessboard_grains(n, m):
    """
    Create an ndarray containing the number of grains on an nxm chessboard.

    Args:
    - n (int): Number of rows on the chessboard.
    - m (int): Number of columns on the chessboard.

    Returns:
    - ndarray: An nxm array with the number of grains on each square.
    """
    total_squares = n * m
    grains = np.array([2**i for i in range(total_squares)])
    chessboard = grains.reshape((n, m))
    return chessboard

# Generate the 8x8 chessboard
chessboard_8x8 = chessboard_grains(8, 8)

# Calculate the sum of grains in the first half (rows 0 to 3) and second half (rows 4 to 7)
first_half_sum = chessboard_8x8[:4, :].sum()
second_half_sum = chessboard_8x8[4:, :].sum()

# Calculate the product of the sums
product_of_halves = first_half_sum * second_half_sum

# Output the results
print(f"Sum of grains in the first half: {first_half_sum}")
print(f"Sum of grains in the second half: {second_half_sum}")
print(f"Product of grains in the two halves: {product_of_halves}")
```

```python
import numpy as np

def chessboard_grains_append(n, m):
    """
    Create an ndarray containing the number of grains on an n x m chessboard using np.append().
    
    Args:
    - n (int): Number of rows on the chessboard.
    - m (int): Number of columns on the chessboard.
    
    Returns:
    - ndarray: An n x m array with the number of grains on each square.
    """
    total_squares = n * m
    grains = np.array([1])  # Start with the first square having 1 grain
    for i in range(1, total_squares):
        grains = np.append(grains, grains[-1] * 2)  # Append double the last value
    return grains.reshape((n, m))


def chessboard_grains_broadcast(n, m):
    """
    Create an ndarray containing the number of grains on an n x m chessboard using broadcasting.
    
    Args:
    - n (int): Number of rows on the chessboard.
    - m (int): Number of columns on the chessboard.
    
    Returns:
    - ndarray: An n x m array with the number of grains on each square.
    """
    total_squares = n * m
    powers = np.arange(total_squares).reshape((n, m))
    grains = 2**powers  # Use broadcasting to calculate 2^i for all positions
    return grains


# Generate an 8x8 chessboard using both methods
chessboard_append = chessboard_grains_append(8, 8)
chessboard_broadcast = chessboard_grains_broadcast(8, 8)

# Results
print("Chessboard generated using np.append():")
print(chessboard_append)

print("\nChessboard generated using broadcasting:")
print(chessboard_broadcast)

# Verify that both methods produce the same results
assert np.array_equal(chessboard_append, chessboard_broadcast), "The two methods should produce the same results."
```

```python
import numpy as np

# Method 1: Using a for loop
def chessboard_grains_for_loop(n, m):
    total_squares = n * m
    grains = []
    current = 1
    for _ in range(total_squares):
        grains.append(current)
        current *= 2
    return np.array(grains).reshape((n, m))

# Method 2: Using np.append()
def chessboard_grains_append(n, m):
    total_squares = n * m
    grains = np.array([1])
    for i in range(1, total_squares):
        grains = np.append(grains, grains[-1] * 2)
    return grains.reshape((n, m))

# Method 3: Using broadcasting
def chessboard_grains_broadcast(n, m):
    total_squares = n * m
    powers = np.arange(total_squares).reshape((n, m))
    return 2**powers

# Set dimensions of the chessboard
n, m = 8, 8

# Timing Method 1: For loop
print("Timing for loop method:")
%timeit chessboard_grains_for_loop(n, m)

# Timing Method 2: np.append()
print("\nTiming np.append() method:")
%timeit chessboard_grains_append(n, m)

# Timing Method 3: Broadcasting
print("\nTiming broadcasting method:")
%timeit chessboard_grains_broadcast(n, m)
```
